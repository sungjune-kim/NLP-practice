# RegEx (Regular Expression)

```python
import re
```

## .

### 한 개의 임의의 문자

```python
r = re.compile("a.c")
r.search("kkk")
```

```python
r.search("apc")
```

-------------------------------------------

## ?

### 앞의 문자가 존재할 수도 있고, 존재하지 않을 수도 있습니다. (문자가 0개 또는 1개)

```python
r = re.compile("ab?c")
r.search("abbc")
```

```python
r.search("abc")
```

```python
r.search("ac")
```

------------------------------------------

## *

### 앞의 문자가 무한개로 존재할 수도 있고, 존재하지 않을 수도 있습니다. (문자가 0개 이상)

```python
r = re.compile("ab*c")
print(r.search("a"))
print(r.search("ac"))
print(r.search("abc"))
print(r.search("abac"))
```

----------------------

## +
### 앞의 문자가 최소 한 개 이상 존재합니다. (문자가 1개 이상)

```python
r = re.compile("ab+c")
print(r.search("a"))
print(r.search("ac"))
print(r.search("abc"))
print(r.search("abdc"))
print(r.search("adbc"))
print(r.search("abac"))
```

----------------------------

## ^
### 뒤의 문자로 문자열이 시작됩니다.

```python
r = re.compile("^a")
print(r.search("abc"))
print(r.search("bc"))
print(r.search("abcd"))
```

-----------------------------------------------------

## {숫자}
### 해당 문자를 숫자만큼 반복한 것을 나타냅니다.

```python
r = re.compile("ab{3}cd")
print(r.search("abbbcd"))
print(r.search("abbcd"))
print(r.search("abbbcde"))
print(r.search("aabbbcdd"))
print(r.search("abbbcad"))
```

--------------------------------------------------

## {숫자1, 숫자2}
### 해당 문자를 숫자1 이상 숫자2 이하만큼 반복합니다

```python
r = re.compile("ab{2,5}c")
print(r.search("abc"))
print(r.search("abbc"))
print(r.search("abbbbbbc"))
print(r.search("abcbbc"))
print(r.search("abbcbbc"))
print(r.search("abbvbbc"))
```

--------------------------------

## {숫자,}
### 해당 문자를 숫자 이상 만큼 반복합니다.

```python
r = re.compile("a{2,}bc")
print(r.search("abc"))
print(r.search("aabc"))
print(r.search("abaac"))
print(r.search("qaaabcd"))
```

----------------------------------

## [ ]
### 문자들 중 한 개의 문자와 매치라는 의미를 가집니다.

```python
r = re.compile("[abc]")
print(r.search("d"))
print(r.search("a"))
print(r.search("abd"))
```

```python
r = re.compile("[a-zA-z]")
print(r.search("320574983y490"))
print(r.search("320574983490"))
print(r.search("320574983S490"))
print(r.search("@#$%#%sFwe2#%"))
```

```python
r = re.compile("[a-z0-9]")
print(r.search("ifiowQFOsokj40d"))
print(r.search("DLJF"))
print(r.search("#@$Jfo-r0"))
```

--------------------------------------------

## [^문자]
### ^ 기호 뒤에 붙은 문자들을 제외한 모든 문자를 매치

```python
r = re.compile("[^a-o]")
print(r.search("asd"))
print(r.search("abcdA"))
print(r.search("as324d"))
print(r.search("uy"))
```

----------------------------------

## re.match()와 re.search()의 차이

```python
r = re.compile("ab.")
print(r.search("kabc"))
print(r.match("kabc"))
print(r.search("abcd"))
```

---------------------------------- 

## re.split()

```python
text = "I really have no idea what is going on with my life."
re.split(" ",text)
```

```python
text = "John/23/Canada/Chef"
re.split("/", text)
```

---------------------------------

## re.findall()

```python
text = "I need 3 onions and 5 tomatos for dinner at 18:00"
print(re.findall("\d+", text))
print(re.findall("\W", text))
```

-------------------------------------

## re.sub()

```python
text = "the national flower of France is rose."
re.sub('[^a-z\W]','{  }',text)
```
